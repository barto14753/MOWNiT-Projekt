# Analiza widma wzajemnego

Celem eksperymentu będzie sprawdzenie czy analiza widma wzajemnego umożliwia odnalezienie korelacji między dwoma szeregami występujących w różnych częstotliwościach. W tym celu wygeneruje sztuczne dane na podstawie dwóch wzorów, które dla pewnych częstotliwości bedą miały silnie zsynchronizowane okresy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
v1 = lambda v0 : np.cos(2 * np.pi * 0.0735 * (v0-1)) + 0.75 * np.sin(2 * np.pi * 0.3 * (v0-1))
v2 = lambda v0 : np.cos(2 * np.pi * 0.0735 * (v0+2)) + 0.75 * np.sin(2 * np.pi * 0.3 * (v0+2))
```

```python
def getData(N, step):
    result = []
    for i in range(N):
        x = i * step
        result.append([x, v1(x), v2(x)])
    return np.array(result)
```

```python
data = getData(1000, 0.01)
print(data)

plt.plot(data[:,0], data[:, 1:] )
plt.show()
```
